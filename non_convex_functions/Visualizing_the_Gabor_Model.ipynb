```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from math import sin, exp
print("imports completed!")
```

## Gabor Model Class

$
\[
f(x, \phi) = \sin\left[\phi_0 + 0.06 \cdot \phi_1 x\right] \cdot \exp\left( - \frac{(\phi_0 + 0.06 \cdot \phi_1 x)^2}{32.0} \right).
\]
$

```python
class GaborModel:

    def __init__(self, phi0: float, phi1:float, x_input: np.array):

        self.phi0 = phi0
        self.phi1 = phi1
        self.x_input = x_input

    def gabor_function(self):

        lin_x = self.phi0 + 0.06 * self.phi1 * self.x_input

        return np.sin(lin_x) * np.exp(- ((lin_x)**2)/32)


```

### Creating the y-outputs using the Gabor Class

```python
### Using Pytorch tensor, not necessary, can be easily done using numpy

## pytorch arange function for the range
x_array = torch.arange(start = -16, end = 16, step = 0.1, dtype = torch.float32)

## converting into numpy array
x_array = x_array.detach().numpy()

## Creating the Gabor class object -> phi_0:-5.0, ph1:25.0
obj1 = GaborModel(phi0 = -5.0, phi1 = 25.0, x_input = x_array)
y1 = obj1.gabor_function()

## Creating the Gabor class object -> phi_0:-5.0, ph1:25.0
obj1 = GaborModel(phi0 = 20.0, phi1 = 40.0, x_input = x_array)
y2 = obj1.gabor_function()

## Creating the Gabor class object -> phi_0:-5.0, ph1:25.0
obj1 = GaborModel(phi0 = 3.0, phi1 = 15.0, x_input = x_array)
y3 = obj1.gabor_function()
```

## Creating a plotting function, to plot the different values of phi0 and phi1

```python
# Plot the function
plt.figure(figsize=(10, 6))

phi0 = -5.0
phi1 = 25.0
plt.plot(x_array, y1, label=f'phi0 = {phi0}, phi1 = {phi1}')

phi0 = 20.0
phi1 = 40.0
plt.plot(x_array, y2, label=f'phi0 = {phi0}, phi1 = {phi1}')

phi0 = 3.0
phi1 = 15.0
plt.plot(x_array, y3, label=f'phi0 = {phi0}, phi1 = {phi1}')


plt.title('Plot of the given function')
plt.xlabel('x')
plt.ylabel('f(x, phi)')
plt.legend()
plt.grid(True)
plt.show()
```
